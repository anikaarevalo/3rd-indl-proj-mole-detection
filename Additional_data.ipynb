### Plotting the distribution of additional synthetic modified data 

```python
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib
import os, sys

#Read in list of files. You might want to look into os.listdir()

path="/Users/anix/3rd-indl-proj-mole-detection/base_dir/train_dir/akiec/"
traces= os.listdir(path)

# Create figure 
fig=plt.figure()
fig.show()
ax=fig.add_subplot(111)

# Grab colormap
cmap = matplotlib.cm.get_cmap('jet')

# Loop through traces and plot them
for j,trace in enumerate(traces):

    # Load file
    dataArray = np.load(path+trace, allow_pickle=True)

    # Grab color
    c=cmap(float(j)/len(traces))

    # Plot
    ax.plot(dataArray.T,color=c)

plt.show()
```
